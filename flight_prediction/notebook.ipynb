```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('datasets/train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('datasets/test.xlsx')
test_df.head()
```

```python
final_df = pd.concat([train_df, test_df])
final_df
```

```python
final_df.info()
```

```python
# final_df['Date_of_Journey'].apply(lambda n: int(n.split('/')[0])) # will convert final_df['Date'] to int type

final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].apply(lambda x: x.split('/')[2])
final_df.head()
```

```python
# Converting date, month, year columns to int type (tho it can be achieved in previous line of code)

final_df['Date'] = final_df['Date'].astype('int64')
final_df['Month'] = final_df['Month'].astype('int64')
final_df['Year'] = final_df['Year'].astype('int64')
final_df.dtypes
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
final_df.head()
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
final_df['Arrival_minute'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
final_df.head()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype('int64')
final_df['Arrival_minute'] = final_df['Arrival_minute'].astype('int64')
final_df.head()
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[0])
final_df['Dep_minute'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[1])
final_df.head()
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Dep_hour'] = final_df['Dep_hour'].astype('int64')
final_df['Dep_minute'] = final_df['Dep_minute'].astype('int64')
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique() # there is only on nan value meaning there is no flight on that route

# final_df[final_df['Total_Stops'] == np.nan]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':2})
final_df.head()
```

```python
# Now we can drop Route as both Route and Total stops give same info

final_df.drop('Route', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
# there are two records containing duration = '5m' which is not possible
# final_df[final_df['Duration']=='5m']
final_df.drop(index=[6474, 2660], axis=0, inplace=True)
final_df[final_df['Duration']=='5m']
```

```python
# Splitting duration field in hour & minute field

final_df['Duration_hour'] = final_df['Duration'].str.split('h').str[0]
final_df['Duration_hour'] = final_df['Duration_hour'].astype('int64')
final_df.head()
```

```python
final_df.drop('Duration', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
# label encoding for categorical features using tranform techniques( better than pd.get_dummies)

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
