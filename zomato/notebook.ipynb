```python
df_country = pd.read_excel('zomato_dataset/Country-Code.xlsx')
df_country
```

```python
df.columns
```

```python
# Combining two dataframes

final_df = df.merge(df_country, how='left', on='Country Code')
final_df.head()
```

```python
# To check datatypes :
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
## Pie Chart - 
plt.pie(country_val, labels=country_names)
```

```python
## Pie Chart - top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.3f%%')
```

Observation : Zomato maximum records or transaction are from India
After that USA and then United Kingdom

```python
### Numerical variable :
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
# Converting it to dataframe
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
# Renaming 0 column to Rating count

ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns={0: "Rating Count"})
ratings
```

### Observation :
1. When rating is between 4.5 to 4.9 --> Excellent
2. From rating is between 4.0 to 4.4 --> very good
3. rating is between 3.5 to 3.9 --> good
4. rating is between 3.0 to 3.4 --> average
5. rating is between 2.5 to 2.9 --> average
6. rating is between 2.0 to 2.4 --> poor

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating color')
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating color', palette=['Pink','red','orange','yellow','green','DarkGreen'])
```

## Observation :
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## Count plot -- plotting with respect to categorical variable
## -- plotting frequency of rating color in ratings dataframe

sns.countplot(x="Rating color", data=ratings,hue='Rating color',palette=['Pink','Red','orange', 'yellow','green','darkgreen'])
```

```python
### Find the countries that has given 0 rating.
final_df[final_df['Aggregate rating']==0.0].groupby('Country').size().reset_index()
```

```python
### Find out which currency is used by which country?
final_df.groupby(['Country','Currency']).size().reset_index().drop(columns=0)
```

```python
# Which countries do have online deliveries??
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts()
```

```python
final_df[['Has Online delivery', 'Country']].groupby(['Has Online delivery', 'Country']).size().reset_index()
```

Observation : Online deliveries are available in India & UAE

#### Create a pie chart for cities distribution :

```python
city_df = final_df.City.value_counts().reset_index(name='Count')
city_df
```

```python
plt.pie(city_df.Count[:5],labels=city_df.City[:5],autopct="%1.2f%%")
```

Observation : Maximum number of transaction are happening in New Delhi then Gurgaon and then Noida, Faridabad and Ghaziabad respectively.
